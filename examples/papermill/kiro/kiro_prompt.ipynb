# Kiro-CLI Atomic Agent Execution

This notebook executes kiro-cli commands via papermill parameterization.
**Security: Uses --trust-all-tools for maximum functionality**

```python
# Parameters cell - will be replaced by papermill
prompt = "Hello! Please introduce yourself."
model = ""
timeout = 300
```

```python
import subprocess
import sys
import os
from datetime import datetime
import json

print("ü§ñ Kiro-CLI Atomic Execution")
print("============================")
print(f"Prompt: {prompt}")
print(f"Model: {model if model else 'default'}")
print(f"Trust Level: ALL TOOLS TRUSTED")
print(f"Timeout: {timeout}s")
print()
```

```python
# Build and execute kiro-cli command
cmd = ["kiro-cli", "chat", "--no-interactive", "--trust-all-tools"]

if model:
    cmd.extend(["--model", model])

cmd.append(prompt)

print(f"Executing: {' '.join(cmd)}")
print()

try:
    result = subprocess.run(cmd, capture_output=True, text=True, timeout=timeout, check=False)
    
    print("üîç Execution Results")
    print("===================")
    print(f"Exit code: {result.returncode}")
    
    if result.stdout:
        print("\nüì§ Kiro-CLI Response:")
        print("-" * 50)
        print(result.stdout)
        print("-" * 50)
    
    if result.stderr:
        print("\n‚ö†Ô∏è  Error Output:")
        print(result.stderr)
        
except Exception as e:
    print(f"‚ùå Execution Failed: {str(e)}")
```
