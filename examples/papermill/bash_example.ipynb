# Bash Kernel Example

This notebook demonstrates using the bash kernel with Papermill.

```bash
# System information
echo "System Information:"
echo "=================="
uname -a
echo
echo "Current directory: $(pwd)"
echo "Current user: $(whoami)"
echo "Date: $(date)"
```

```bash
# File operations
echo "Creating sample files..."
echo "Hello from Papermill!" > sample.txt
echo "Line 2" >> sample.txt
echo "Line 3" >> sample.txt

echo "File contents:"
cat sample.txt

echo
echo "File info:"
ls -la sample.txt
```

```bash
# Process information
echo "Process information:"
echo "==================="
ps aux | head -10

echo
echo "Memory usage:"
free -h
```

```bash
# Cleanup
echo "Cleaning up..."
rm -f sample.txt
echo "Done!"
```
